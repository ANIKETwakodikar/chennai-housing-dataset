```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import gradio as gr
import requests
import io
```

```python
url = "https://raw.githubusercontent.com/ANIKETwakodikar/chennai-housing-dataset/main/Chennai.csv" 
download = requests.get(url).content
df = pd.read_csv(io.StringIO(download.decode('utf-8')))

af=df.drop(columns=['Price', 'Location', 'Resale',"24X7Security",
       'MaintenanceStaff', 'Gymnasium', 'LandscapedGardens',
       'JoggingTrack', 'RainWaterHarvesting', 'IndoorGames', 'ShoppingMall',
       'Intercom', 'SportsFacility', 'ATM', 'ClubHouse', 'School', 'PowerBackup', 'CarParking', 'StaffQuarter',
       'Cafeteria', 'MultipurposeRoom', 'Hospital', "Children'splayarea", 'VaastuCompliant', 'Microwave', 'GolfCourse',
       'DiningTable', 'Wardrobe'])
```

```python
model=linear_model.LinearRegression()
model.fit(af,df['Price'])
af
```

```python
def tello(area,no_of_bedrooms,swimming_pool,Washing_Machine,Gas_connection,AC,wifi,lift,beds,TV,sofa,fridge):
    a=int(model.predict([[area,no_of_bedrooms,swimming_pool,Washing_Machine,Gas_connection,AC,wifi,lift,beds,TV,sofa,fridge]]))
    if a>10000000:
        return str(a//10000000)+' Cr'
    else:
        return str(a//100000)+' L'
```

```python
model_gr=gr.Interface(tello,inputs=["number","number","number","number","number","number","number","number","number","number","number","number"],outputs='text')
model_gr.launch()
```
